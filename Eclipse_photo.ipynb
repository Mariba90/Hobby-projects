```python
image = cv.imread(path+"eclipse.jpeg")
h, w = image.shape[:2]

half_height, half_width = h//4, w//8
transition_matrix = np.float32([[1, 0, half_width],
                               [0, 1, half_height]])

transition_matrix_1 = np.float32([[1, 0, -600],
                               [0, 1, -300]])

img_transition = cv.warpAffine(image, transition_matrix, (w, h))
img_transition_1 = cv.warpAffine(image, transition_matrix_1, (w, h))


img_transition = cv.warpAffine(image, transition_matrix, (w, h))
plt.imshow(cv.cvtColor(img_transition+img_transition_1, cv.COLOR_BGR2RGB))
plt.title("Translation")
plt.show()
```
